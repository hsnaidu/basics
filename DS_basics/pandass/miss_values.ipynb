HANDLING MISSING VALUES

1. fillna()
2. fillna() with different values
3. Interpolate : Guess onto missing values using interpolation
4. dropna, drop-rows, with missing values

```python
df = pd.read_csv("C:\\Users\\harip\\Desktop\\refrence\\data_science\\dataset\\weather_data.csv")
df.head()
```

```python
df = pd.read_csv("C:\\Users\\harip\\Desktop\\refrence\\data_science\\dataset\\weather_data.csv",parse_dates = ['day'])
df.head()
```

* When you use Inplace = True, The original dataset wont be changed insted a copy of the dataframe is returned

```python
df.set_index('day',inplace=True)
df.head()
```

* We often REPLACE the NA values with something meaningfull values

```python
# NOTE fillna()
df.head()

# filling the NA values
ndf  = df.fillna(0)
ndf # NOTE All the NA values are filled with 0
```

* Use the concept of dict{ } to fill with custom values 

```python
new_df = df.fillna({
    # NOTE It doesnt work with the filled NA values
    'temperature' : 1,
    'windspeed' : 0,
    'event' : 'no-event'
})
new_df
```

FILL WITH NA VALUES USING "ffill" and "bfill"

```python
df.head()
```

```python
# NOTE 1. FORWARD FILL
f_fill = df.fillna(method='ffill')
f_fill.head()
```

```python
# NOTE 2. BACKWARD FILL
b_fill = df.fillna(method='bfill')
b_fill.head()
```

* YOU CAN DO THIS SAMETHING USING TO COLUMNS AS WELL

```python
new_df = df.fillna(method="backfill",axis='columns')
new_df.head() # NOTE carefull while using categorical values
```

* WE CAN LIMIT THE NUMBER OF TIMES THE VALUES IS COPIED TO ANOTHER CELL

```python
# NOTE LIMTI allows you can only copy once
new = df.fillna(method='ffill',limit=1)
new # Dosen't copy more than one time the same number 
```

* INTERPOLATE METHOD

Case Scenario : Let's say the temp on 1/1/2017 was 32 and the temp on 1/5/2017 is 28, but the temp on 1/4 can be btw those two days so we use this method for that  

CURVE FITTING USING LINEAR POLYNOMIALS TO CONSTRUCT A NEW DATA POINTS

```python
df
```

```python
df = df.interpolate()
df
```

You have different methods in interpolate

```python
new_df = df.interpolate(method='cubic')
new_df = df.interpolate(method='time')
new_df = df.interpolate(method='linear')
new_df # NOTE This interpolate values are based on time 
```

* dropna()

1. When ever you have Less Data and Medical Data -> Don't drop the values 

```python
df
```

```python
# here we are dropping the null values
null_df = df.dropna()
null_df
```

ADDING CONDITION TO DROP FUNCTION

```python
# NOTE drop the rows only if all the values are null
all_df = df.dropna(how='all')
all_df
```

```python
# NOTE THRESH
# If a rows has 2 valid values keep that row
two_keep = df.dropna(thresh=2)
two_keep
```
