REPLACE METHOD FOR MISSING DATA

```python
df = pd.DataFrame({
    'day' : pd.date_range("01-01-2017","01-7-2017"), # Series of Continues Dates
    'temp' : [32,-9999999,23,-9999999999999,32,32,21],
    'windspeed' : [6,7,-9999,7,-9999,2,5],
    "event" : ['Rain',"Snow","Sunny","No Event","Rain","Snow","No Event"]
})
df
```

```python
# NOTE replace the -9999999 numbers
df.replace(-9999999999999,1) # Replacing one value
```

REPLACING WITH ANOTHER LIST OF VALUES

```python
df = pd.DataFrame({
    'score' : ['exceptional','average','good','poor','average','exceptional'],
    'student':['rob','maya','parthiv','tom','julian','erica']
})
df
```

REPLACING WITH A LIST 

```python
ndf = df.replace(['poor','average','good','exceptional' ],[0,1,2,3])
ndf
```
