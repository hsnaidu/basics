```python
import pandas as pd
```

```python
india = pd.DataFrame({
    "city" : ['bangalore','delhi','chennai','hydrabadh'],
    "temp" : [23,23,34,54]
})
usa = pd.DataFrame({
    "city" : ['newyork','chinatown','orlando','asutralia'],
    "temp" : [34,32,34,56]
})
```

```python
df = pd.concat([india,usa])
df
```

SEE THE ORDER OF SN NUMBER : WE NEED A CONTINUES NUMBER 

```python
# NOTE we just ignore the index 
df = pd.concat([india,usa],ignore_index=True)
df
```

SIMILAR TO GROUPING WE CAN GROUP THE DATA USING THE DATAFRAME NAME

```python
df = pd.concat([india,usa],keys = ['india','usa'])
df # NOTE you get two different col of india and usa 
```

We can add more col similar way

```python
temp =pd.DataFrame({
    "temp_two" : [23,23,34,54,34,56]
})
temp
```

```python
pd.concat([df,temp]) # NOTE re-check this once 
```
