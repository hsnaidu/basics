```python
# NOTE Matplotlib Basics

# NOTE NOTE https://matplotlib.org/2.0.2/gallery.html
import numpy as np
import matplotlib.pyplot as plt
```

* They are used to check the co-rellation and how the data is spread
* Data Points plotted as dots / points



```python
x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100

# print(x_data,y_data)
```

SCATTER PLOT

```python
# NOTE syntax
plt.scatter(x_data,y_data) # PASSING DATA
plt.show() # SHOWING THE GRAPH'S [Not impt]
```

COLOR

```python
plt.scatter(x_data,y_data,c = 'red') 
# NOTE you can also provide hexa-code plt.scatter(x_data,y_data,c = "#76GGH")
```

MARKERS : Changing the pattern

```python
plt.scatter(x_data,y_data,c="purple",marker="*")
```

ALPHA [TRANSPARENCY] : Used when we have overlapping values 

```python
x_data = np.random.random(1000) * 100
y_data = np.random.random(1000) * 100
plt.scatter(x_data,y_data,c="purple",marker="*")
```

```python
x_data = np.random.random(1000) * 100
y_data = np.random.random(1000) * 100
plt.scatter(x_data,y_data,c="purple",marker="*",alpha=0.4)
```
