```python
# NOTE Matplotlib Basics

# NOTE NOTE https://matplotlib.org/2.0.2/gallery.html
import numpy as np
import matplotlib.pyplot as plt
```

* Time series data
* You need to keep track of an object over time

```python
years = [2006 + x for x in range(14)] # basically adding up years from 2006 + 16years

weight  = [84,85,86,76,79,83,79,83,80,83,81,79,78,87]
# NOTE we are tracking a person weight's
```

LINE CHART

```python
plt.plot(years, weight)
plt.show()
```

COLOR

```python
plt.plot(years, weight, color = 'g')
plt.show()
```

LINE WIDTH : HOW THICK YOU WANT YOUR LINE TO BE

```python
plt.plot(years, weight, color = 'g', lw = 3)
plt.show()
```

ADDING A LINE STYLE

```python
plt.plot(years, weight, color = 'g', lw = 3, linestyle = "--")
plt.show()
```

SHORT CUT : Instead of mentioning all the parameters 

```python
plt.plot(years, weight,'r--')
plt.show()
```
