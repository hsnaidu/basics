WE OFTEN REPLACE THE NA VALUES WITH SOMEMEANING FULL VALUES OF A GUESS 

1. fillna()

```python
new_df = df.fillna(0)
new_df
# everything that was na is now 0,  and always having zero is not a good option so we need to comeup with a new data
```

```python
# new_df = df.temperature.fillna(1) you will only get the temprature value 
# so we are using a concept of dictionary to fill the multiple values with different values
new_df = df.fillna({ # it dosent work with the filled na values
    'temperature' : 1,
    'windspeed' : 0,
    'event' : 'no event'
})
new_df
```

TO FILL THE NA VALUES WE NEED TO COME UP WITH SOME ESTIMATE

1. FORWARDFILL / BACKWARDFILL

```python
new = df.fillna(method='ffill') 
new # carry forward the values to the next 
new_df = df.fillna(method='bfill')
new_df
# more on the pandas docs
# similarly you have axis
new_df = df.fillna(method="backfill",axis='columns')
new_df # need to be carefull when we use this casue it will mix categorical values

# we have something called LIMTI you can only copy once
new = df.fillna(method='ffill',limit=1)
new # it dosent copy more than one time the same number 
```

INTERPOLATE METHOD : why : lets say the temp on 1/1 was 32 and the temp on 1/5 is 28 , but the temp on 1/4 can be btw those two days, or may be not but we have a method for that

curve fitting using linear polynomials to construct new data points 

```python
new_df = df.interpolate() # comes with the linear interpolate like a better guess
new_df # it tries to find the middle values based on the linear interpolation
```

```python
# there are so many other method we can mention in methods like cubic,time and stuff , more on pandas org 
new_df = df.interpolate(method='time')
new_df # this is based on time TIME NOTE NOTE NOTE this is dealing with time 
# analogy in day - month there is 1 and 4 , 2 and 3 are missing so the temp can vary in those two days as well so we use time 
```

* NOTE : WHEN YOU HAVE LESS DATA AND MEDICAL DATA DO NOT DROP THE VALUES 
* HERE WE ARE DROPPING THE NULL VALUES 


2. dropna()

```python
# dropping all the rows with na values 
new_df = df.dropna()
new_df  # the dataset size went from 9rows to 3rows so think when you want to drop 
```

I NEED TO DROP SOME ROWS BUT I ALSO NEED TO PRESERVE THEM BECAUSE THEY HAVE SOME VALUES

```python
# IF ALL THE ROWS HAVE NA THEN ONLY DROP
new_df = df.dropna(how='all')
new_df # only the rows which has all the null values are dropped

# more conditions 
# if i have atleast 1 non na values keep that row
new_df = df.dropna(thresh=2) # i ned only rows which has 2 valid value's
new_df
```

IN THE ABOVE DATA WE DONT HAVE A CONTINUES DATE SO WE USE DatetimeIndex and reindex it back

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df # so we have second and third rows no we can use fill na values to get some values
new_df = df.interpolate(method='time') # if you dont specify the method it takes linear interpolation
new_df # all new values are filled based on time 
new_df.fillna(method='ffill',inplace=True)
new_df
```

3. USING REPLACE METHOD WITH MISSING DATA

```python
new_df = pd.DataFrame({
    'day' : pd.date_range("01-01-2017","01-7-2017"), # for continues data
    'temp' : [32,-9999999,23,-9999999999999,32,32,21],
    'windspeed' : [6,7,-9999,7,-9999,2,5],
    "event" : ['R',"S","SN","No Event","R","S","No Event"]
})
new_df
```

```python
new_df = new_df.replace([-9999,-9999999,-9999999999999,'No Event'],np.NaN) # yeah so you can replace multiple values and also different values 
# now i can perform the filling method 
# but if we have No Event in other field which is valid it will change the entire dataset so use dictionary method to change the values
new_df = new_df.replace({
    'event' : 'No_event' # no_event is replaced by nan , or any other value
},np.NaN)
new_df
```

# Replacing a List of values with another list of values

# Concat Dataframes

JOIN TWO OR MORE DATAFRAME

```python
# lets have weather data saved of two-diff country
india = pd.DataFrame({
    "city" : ['ba','dl','cn','tn'],
    "temp" : [23,23,34,54]
})
usa = pd.DataFrame({
    "city" : ['ny','chi','orl','asu'],
    "temp" : [34,32,34,56]
})
```

```python
# NOTE WE HAVE CREATE A TWO DATA FRAME NOW WE JUST NEED TO JOIN IT 
main = pd.concat([india,usa])
main
```

```python
# NOTE IF WE NEED A CONTINUES INDEX 
main = pd.concat([india,usa],ignore_index=True)
main # now we have a continues index numbers
```

NOTE LET'S SAY I WANT TO RETRIVE WEATHER DATA OF INDIAN CITIES SIMILAR TO GROUPING 

```python
df = pd.concat([india,usa],keys = ['india','usa'])
df # NOTE you get two different col of india and usa 
```

NOW I NEED temp_two as one more col : I can specify it to be joined as a row or a column
