# Operations with Dataframe - Impt

```python
df = pd.read_csv('C:\\Users\\harip\\Desktop\\refrence\\data_science\\dataset\\nyc_weather.csv')
df.head(5)
```

```python
# To find the max value
df.Temperature.max()

# To find the min value
df.Temperature.min()

# To find the mean value 
df.Temperature.mean()

# To find the standard deviation
# They measure how close - far the data points are from eachother 
# https://gemini.google.com/app/22c12f42dadf56be
df.Temperature.std()
```

TO GET ALL THE STATSTICAL MEASURE'S

```python
round(df.describe(),1)
```

CONDITIONAL SELECT IN DATAFRAME

```python
df['Temperature']>=32 # will only give temp value in boolean value 
df[df.Temperature >= 32] # this will give out the entire dataframe
df[df.Temperature == df.Temperature.max()] # this only print the rows the temprature with max value 

# IF YOUR COL-NAME CONTAIN SPACES THEN YOU HAVE TO USE 
df[df.Temperature == df['Temperature'].max()]

# lets say you only want a specific col when the temp is MAX
df['EST'][df.Temperature == df['Temperature'].max()]

# similarly you can also print multiple col
df[['EST','Humidity']][df.Temperature == df['Temperature'].max()]
```

SET INDEX - LOC and ILOC

```python
df.index 
# lets say you want to change the index to some-other value 
df.set_index('EST')
```

TO RESET THE INDEX BACK TO ORIGINAL df.reset_index(inplace = True)

The use of inplace is basically to apply the changes and not return anything 
[video on how it works](https://www.youtube.com/watch?v=XaCSdr7pPmY)

```python
sdf = df.head()
sdf = sdf.set_index('EST')
sdf
```

```python
# NOTE This loc is for specified value when you know what to search in the data
sdf
sdf.loc['1/1/2016']
sdf.loc['1/3/2016']
```
