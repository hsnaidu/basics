```python
# NOTE Matplotlib Basics

# NOTE NOTE https://matplotlib.org/2.0.2/gallery.html
import numpy as np
import matplotlib.pyplot as plt
```

BAR CHARTS 

```python
programming_x = ["C++","JAVA","PYTHON","C#","GOLANG","R"]
people_y = [100,120,500,230,20,160]
```

```python
plt.bar(programming_x,people_y)
# plt.plot(programming_x,people_y, 'r--', lw=2)
```

COLOR

```python
plt.bar(programming_x,people_y,color = 'red') # NOTE we need to specify it as color 
```

ALIGINING

```python
plt.bar(programming_x,people_y,color = 'red', align='edge') # NOTE changes the bottom reference line from middle to side
```

WIDTH OF THE BAR

```python
plt.bar(programming_x,people_y,color = 'red', width=0.1) 
# we can even change the border color / edege color
plt.bar(programming_x,people_y,color = 'red', width=0.6, edgecolor = 'blue') 
plt.bar(programming_x,people_y,color = 'red', width=0.6, edgecolor = 'blue', lw = 2) 
```
