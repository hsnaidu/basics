##### Line Graph : Example

```python
# 1-d graph
a = [1,2,3,4]
b = [11,12,13,14]

# you can set your own size and dpi : NOTE ALWAYS AT THE BEGINING
# plt.figure(figsize=(10,15), dpi=300) 
plt.figure(figsize=(5,5))

plt.plot(a,b, label = '+10') # we need to add the lable element for legened 


# To add a graph title
plt.title('A simple line graph') # Title
# To add lables on axis
plt.xlabel(" X Axis Lable")
plt.ylabel(" Y Axis Lable")

# You can also play with the Font 
plt.title('Simple Line Graph',fontdict={'fontname' : 'Times New Roman','fontsize':20}) # font changed and font-size changed # NOTE not that important

# What if we dont want an decimal place number
plt.xticks(a)
plt.yticks(b)

# NOTE NOTE NOTE Graph will resize automatically based on the number values given 

# adding a info, what is the line denoting 
plt.legend()

# Line style
# you can also pass hexa decimal number 
plt.plot(a,b, label = '+10' ,color = 'g', linewidth = 5, marker = '*', markersize = 20, markeredgecolor = 'blue') 

# changing the line style
plt.plot(a,b, label = '+10', linestyle = '--') 

# There is a short hand technique to do all this
# fmt = [color][marker][line]
plt.plot(a,b,'b^--', label = '10+')

# NOTE NOTE if we want to save our graph
plt.savefig('mygraph.png',dpi=300)




```

```python
# Plotting using the numpy  # lil complex to understand 
a = np.arange(0,6,1)
# plot one more line square of it 
# plt.plot(a**2,'r.-',label='Square')
# plt.plot(a**3,'g*-',label = 'Cube')

# now i need the cube after 3 in dotted 
plt.plot(a[:5],a[:5]**3,'g',label = 'Cube')
plt.plot(a[4:],a[4:]**3,'r--') # from 
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

##### Bar Chart : Example

```python
lable = ['a','b','c','d']
values = [1,2,3,4]
# They follow the this formate (Y,X)
# same properties as line graph applies
plt.bar(lable,values)
plt.title('Simple Bar Graph',fontdict={'fontname' : 'Times New Roman'})

# https://youtu.be/3Xc3CA655Y4?t=1824
# Adding styles to bar chart 
plt.show() # removes the unwanted information
```
